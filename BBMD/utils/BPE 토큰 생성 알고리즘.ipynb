# BPE 생성

```python
# TheList = []
filePath = '/content/drive/MyDrive/Capstone Design Project + 한이음 과정/code/BPE 알고리즘2/vocab_checkpoint.p'
with open(filePath, 'rb') as lf:
    TheList = pickle.load(lf)
    print(TheList)
```

```python
len(TheList)
```

```python
for ind in tqdm.notebook.tqdm(range(100)):
    tempList = {}
    tempList2 = {}

    for opList in tqdm.notebook.tqdm(opcodeList):
        for i in range(len(opList) - 1):
            tempPair = opList[i]+opList[i+1]
            

            tempPairStr = ''
            for j in range(len(tempPair)):
                tempPairStr = tempPairStr + tempPair[j]
            
            tempList.setdefault(tempPairStr, 0)
            tempList2.setdefault(tempPairStr, opList[i] + opList[i+1])
            tempList[tempPairStr] += 1

    maxW = max(tempList,key=tempList.get)
    TheList.append(tempList2[maxW])

    for opList in tqdm.notebook.tqdm(range(len(opcodeList))):
        for i in range(len(opcodeList[opList])):            
            if len(opcodeList[opList]) <= i:
                break

            if tempList2[maxW][0] != opcodeList[opList][i][0]:
                continue

            tempL = []
            lenC = 0

            for j in range(len(tempList2[maxW])):
                if i+j >= len(opcodeList[opList]):
                    break
                tempL = tempL + opcodeList[opList][i+j]
                lenC += len(opcodeList[opList][i+j])
                if lenC >= len(tempList2[maxW]):
                    break
            if len(tempL) > len(tempList2[maxW]):
                continue

            if tempL == tempList2[maxW]:
                sL = len(tempList2[maxW]) - len(opcodeList[opList][i])
                opcodeList[opList][i]=tempList2[maxW]

                while True:
                    del opcodeList[opList][i+1]
                    if len(opcodeList[opList]) <= i+1:
                        break

                    sL -= len(opcodeList[opList][i+1])
                    if sL <= len(tempL):
                        break

    rand_num = random.randrange(0, len(oL)-20)

    if len(TheList) % 10 == 0:
        with open('/content/drive/MyDrive/Capstone Design Project + 한이음 과정/code/BPE 알고리즘2/vocab_checkpoint.p', 'wb') as file:
            pickle.dump(TheList, file)  
        with open('/content/drive/MyDrive/Capstone Design Project + 한이음 과정/code/files/BPE Database/opcodeList_checkpoint.p', 'wb') as file:
            pickle.dump(opcodeList, file)
    
    if len(TheList)%500 == 0:
        name = '/content/drive/MyDrive/Capstone Design Project + 한이음 과정/code/BPE 알고리즘2/' + str(len(TheList)) + 'vocab.p'
        with open(name, 'wb') as file:
            pickle.dump(TheList, file)
        name = '/content/drive/MyDrive/Capstone Design Project + 한이음 과정/code/files/BPE Database/' + str(len(TheList)) + 'opcodeList_checkpoint.p'
        with open(name, 'wb') as file:
            pickle.dump(opcodeList, file)  

    print('-----------------------------------------------------------')
    print(f'Vocab크기: [BPE 토큰개수, {len(TheList)}] + [opcode 개수{len(opcodes)}] = {len(TheList)+len(opcodes)}')
    print(f'{tempList2[maxW]} - > {maxW}')
    print(f'빈도수: {tempList[maxW]}')

    for i in range(4):
        print(f'(Random)변환 이전: {oL[rand_num+i]} -> 변환 이후: {len(opcodeList[rand_num+i])}')
```
